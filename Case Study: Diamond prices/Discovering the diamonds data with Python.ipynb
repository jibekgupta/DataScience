```python
# Import packages
import numpy as np
import pandas as pd
import seaborn as sns

# Import data and explore data
diamonds = pd.read_csv('/content/diamonds_casestudy.csv')
diamonds.head()
```

```python
# Describe the stats
diamonds.describe()
```

```python
# Calculate group means based on cut
diamonds.groupby(by=['cut'])[['carat','depth','table','price','width','length','height']].mean()
```

```python
# Calculate group sizes for cut
diamonds.groupby(by=['cut']).size()
```

```python
# Calculate group means by color
diamonds.groupby(by=['color'])[['carat','depth','table','price','width','length','height']].mean()
```

```python
# Calculate group sizes by color
diamonds.groupby(by=['color']).size()
```

```python
# Pivot table with average price for cut and color combinations
diamonds.pivot_table(values='price',index='color',columns='cut',aggfunc='mean')
```

```python
# Pivot table with group sizes for cut and color combinations
diamonds.pivot_table(values='price', index='cut', columns='color', aggfunc='size')
```
