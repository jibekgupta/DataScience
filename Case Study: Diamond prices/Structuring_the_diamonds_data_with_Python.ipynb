```python
# Import package and data
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn import preprocessing

diamonds = pd.read_csv('/content/diamonds_casestudy.csv')
diamonds.head()
```

```python
# STandardize prices
prices = diamonds['price']
std_price = preprocessing.scale(prices)
diamonds['std_price'] = std_price
diamonds.head()
```

```python
# Histogram of prices
plt.hist(prices, color='red')
plt.xlabel('Prices', fontsize=14, color='red')
plt.ylabel('Count', fontsize=14, color='red')
plt.title('Diamond prices', fontsize=14, color='red')
plt.show()
```

```python
# Histogram of standardized prices
plt.hist(std_price, color='blue')
plt.xlabel('Standardized prices', fontsize=14, color='blue')
plt.ylabel('Count',fontsize=14, color='blue')
plt.title('Diamond prices', fontdict={'fontsize': 14, 'color':'blue'})
plt.show()
```

```python
# Scatterplot of width and length
plt.scatter(x=diamonds['width'], y=diamonds['length'], color='green')
plt.xlabel('Width (mm)', fontsize=14, color='green')
plt.ylabel('Length (mm)', fontsize=14, color='green')
plt.title('Diamond dimensions', fontdict={'fontsize': 14, 'color':'green'})
plt.show()
```

```python
# STandardized table and produce a plot
table = diamonds['table']
std_table = preprocessing.scale(table)
plt.hist(std_table, color='yellow')
plt.xlabel('Standardized table', fontdict={'fontsize':14, 'color':'yellow'})
plt.ylabel('Count', fontdict={'fontsize':14, 'color':'yellow'})
plt.title('Diamond table', fontdict={'fontsize':14, 'color':'yellow'})
plt.show()
```

```python
# Plot price, width, and table
# Plot.subplot(rows,columns, plot index)
plt.figure(figsize=(19,5))

plt.subplot(1,3,1)
plt.hist(diamonds['price'])
plt.xlabel('Prices (USD)', fontsize=12)
plt.ylabel('Count', fontsize=12)

plt.subplot(1,3,2)
plt.hist(diamonds['width'])
plt.xlabel('Width (mm)', fontsize=12)
plt.ylabel('Count', fontsize=12)

plt.subplot(1,3,3)
plt.hist(diamonds['table'])
plt.xlabel('Table', fontsize=12)
plt.ylabel('Count', fontsize=12)

plt.show()
```
